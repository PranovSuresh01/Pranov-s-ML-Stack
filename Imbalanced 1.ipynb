```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
range_gamma = np.linspace(0.3,0.5, 10)
range_P = []
range_R = []
for i in range_gamma: 
    P = i/(2*i-0.3)
    R = 2*i
    range_P.append(P)
    range_R.append(R)
```

```python
plt.plot(range_gamma, range_P, label ="Precision")
plt.plot(range_gamma, range_R, label ="Recall")
plt.xlabel("Gamma")
plt.ylabel("Value")
plt.legend()
plt.title("Precision and Recall as functions of Gamma")
plt.plot()
```
